```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
```

```python
dado = np.loadtxt(url, delimiter=',', skiprows=1, usecols=np.arange(1,6,1))
```

```python
diametro_laranja = dado[:5000,0]
diametro_toranja = dado[5000:,0]
peso_laranja = dado[:5000,1]
peso_toranja = dado[5000:,1]
```

```python
import matplotlib.pyplot as plt

plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
```

```python
Y = peso_laranja
X = diametro_laranja
n = np.size(peso_laranja)

Z = peso_toranja
W = diametro_toranja
m = np.size(peso_toranja)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
c = (m*np.sum(W*Z) - np.sum(W)*np.sum(Z))/(m*np.sum(W**2) - np.sum(W)**2)
```

```python
b = np.mean(Y) - a*np.mean(X)
d = np.mean(Z) - c*np.mean(W)
```

```python
y = X*a + b
z = W*c + d
```

```python
plt.plot(diametro_laranja, peso_laranja)
plt.plot(X,y)
```

```python
plt.plot(diametro_toranja, peso_toranja)
plt.plot(W,z)
```
